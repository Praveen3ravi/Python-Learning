```python
data1 = pd.read_csv('college_1.csv')
df=np.nan_to_num(data1,nan=0)
df1=pd.DataFrame(data1)
data2 = pd.read_csv('college_2.csv')
df=np.nan_to_num(data2,nan=0)
df2=pd.DataFrame(data2)
data_merge = pd.merge(df1, df2, how="outer")
col=["Name","python","mysql","Previous Geekcoins","CodeKata Score","Department","Rising","python_en","computational thinking"]
training_data=pd.DataFrame(np.array(data_merge),columns=col)
training_data
```

```python
super=np.array(training_data,copy=True)
super=np.nan_to_num(super,nan=0)
col=["Name","python","mysql","Previous Geekcoins","CodeKata Score","Department","Rising","python_en","computational thinking"]
test1=pd.DataFrame(super,columns=col)
test1=test1[test1['CodeKata Score']>15000]
np.savez("Exceeded expectations.csv",test1)
test1
```

```python
good=np.array(training_data,copy=True)
good=np.nan_to_num(good,nan=0)
col=["Name","python","mysql","Previous Geekcoins","CodeKata Score","Department","Rising","python_en","computational thinking"]
test2=pd.DataFrame(good,columns=col)
test2=test2[(test2["CodeKata Score"]>10000) & (test2["CodeKata Score"]<15000)]
np.savez("Reached_expectations.csv",test2)
test2
```

```python
dept=np.array(training_data,copy=True)
col=["Name","python","mysql","Previous Geekcoins","CodeKata Score","Department","Rising","python_en","computational thinking"]
dept=pd.DataFrame(dept,columns=col)
g=dept.groupby("Department")
one=g.get_group("Computer Science and Engineering")
two=g.get_group("Electronics and Electrical Engineering")
third=g.get_group("Electronics and Communication Engineering")
CSE=np.amax(one["CodeKata Score"])
EEE=np.amax(two["CodeKata Score"])
ECE=np.amax(third["CodeKata Score"])
g.max()
```

```python
val=[CSE,EEE,ECE]
lab=['CSE','EEE','ECE']
plt.axis("equal")
plt.title("Department wise codekata performence",bbox={'facecolor':'0.8', 'pad':10})
plt.pie(val,labels=lab,radius=1,autopct="%0.1f%%")
plt.show()
```
